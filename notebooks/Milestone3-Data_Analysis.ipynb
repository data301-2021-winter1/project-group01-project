Research Question: How do pro esports competitions streamed on media sites such as Twitch affect the peak average player base in a game like CSGO?

The plot I used is an x - Month and y - Average players and Average viewers to show the effect of esports competition streams streamed on Twitch.

Counter Strike: Global Offensive esports tournaments and competitions span from March to April and September to December. In March, where every team that can sign up will play at playoffs, the matches start. Then when April comes, that is when the teams who won more games than other teams get to progress through the semi-finals, and that is where more viewers come to watch the Twitch streams. The average viewer data shows from the spike in average viewership around the month of April (4th month). The people want to watch the games when the semi-finals and finals start because the popular teams fight against each other.
The more viewership a game has, the more reason it is for people to play it because of how popular it is. This reason shows from the data of the average player base. During the month of April, the average player base increased around 200000 more players. After the tournament, the viewership dies down a bit because the excitement also dies down, and then the viewership increases again around September because of another big competition. Since the last tournament, there were a lot of players already playing so it makes sense that the increase is lesser than the April increase but it is still increasing because of the continuous tournaments.
