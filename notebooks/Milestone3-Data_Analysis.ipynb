From the steam popularity dataset, I extracted the rows about Counter-Strike: Global Offensive, the most popular game on Steam, to process the data and wrangle the average players, gained average players from the previous month, and the seven-month rolling gain average for every month recorded in the dataset. This allows us to have a timeline of the average, the gain, and the seven-month rolling gain average. All we need to do with these timelines is see if there are any dips in the average players that are not a part of a longer pre-existing trend and compare them to the release dates of triple-A titles like Call of Duty, Battlefield, Halo, Overwatch, etc.
